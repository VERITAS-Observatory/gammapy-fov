```python
from gammapy_bkg2 import *
from gammapy.maps import MapAxis
from gammapy.datasets import MapDataset
from gammapy.makers import (MapDatasetMaker, SafeMaskMaker, DatasetsMaker, FoVBackgroundMaker)
import yaml
```

```python
datadir = '/nevis/tehanu/data/st3624/software/gammapy-fov/data'
dl3dir = '/nevis/tehanu/data/st3624/software/gammapywork/dl3_fullenclosure_moderate2tel/dl3/'
srcdict = f'{datadir}/CTA1_ep_wob_dur_el_az_cur.json'
with open(srcdict, 'r') as fp: srcinfo = json.load(fp)
```

```python
srcinfo_na = [info for info in srcinfo if info['epoch']=='na']
srcinfo_ua = [info for info in srcinfo if info['epoch']!='na']
```

```python
len(srcinfo_ua)
```

```python
egdict_suffix = '_good_egruns_dur_el_az_cur_srcid.json'
egdict_ua = [datadir+'/'+file for file in os.listdir(datadir) if egdict_suffix in file]
egdict_na = [egdict for egdict in egdict_ua if 'na' in egdict]
egdict_ua = [egdict for egdict in egdict_ua if 'oa' not in egdict and 'na' not in egdict]
```

```python
eginfo_ua = []
for file in egdict_ua:
    with open(file, 'r') as fp: eginfo_ua += json.load(fp)
```

```python
eginfo_na = []
for file in egdict_na:
    with open(file, 'r') as fp: eginfo_na += json.load(fp)
```

```python
with open('analysis_config.yml', 'r') as f:
    config = yaml.safe_load(f)
```

```python
config
```

```python
ds = DataStore.from_dir(dl3dir)
srcobs_na = get_available_obs(ds, [run['run_id'] for run in srcinfo_na])
srcobs_ua = get_available_obs(ds, [run['run_id'] for run in srcinfo_ua])
```

```python
plot_runs_info(srcinfo_na,elmin=40,elmax=50,thresh_az=config['mimic_data']['az_thresh'])
```

```python
plot_runs_info(srcinfo_ua,elmin=40,elmax=50,thresh_az=config['mimic_data']['az_thresh'])
```

## NA runs

```python
mimic_na = MimicDatasetMaker(ds, srcinfo_na, eginfo_na, az_thresh=config['mimic_data']['az_thresh'])
```

```python
np.sum([info['duration'] for info in srcinfo_na])/60/60
```

```python
mimic_na.find_mimic_runs(times=2)
```

```python
plot_runs_info(mimic_na.mimic_datasets[0]['information'],elmin=40,elmax=60)
```

```python
mimic_na.run_blank_sky(0)
```

```python
bkg_width = config['acceptance']['width']
bkg_nbin = int(bkg_width / config['acceptance']['binsz'])
fov_lon = MapAxis.from_bounds(-bkg_width/2, bkg_width/2, nbin=bkg_nbin, interp="lin", unit="deg", name="fov_lon")
fov_lat = MapAxis.from_bounds(-bkg_width/2, bkg_width/2, nbin=bkg_nbin, interp="lin", unit="deg", name="fov_lat")
```

```python
energy_binning = config['energy_binning']
energy_axis = MapAxis.from_energy_bounds(
    energy_binning['energy_lo']*u.TeV, energy_binning['energy_hi']*u.TeV, energy_binning['energy_bins'])
```

```python
bkgmaker = Background3DModelEstimator(energy_axis, fov_lon, fov_lat)
bkgmaker.run(mimic_na.mimic_datasets[0]['observations'])
```

```python
bkgrate = bkgmaker.background_rate
```

```python
bkgrate.peek()
```

```python
plot_bkg(bkgrate)
```

```python
for idx, data in enumerate(bkgrate.data):
    bkgrate.data[idx] = convolve_nonzero(data)
```

```python
plot_bkg(bkgrate)
```

```python
bkgrate.write('CTA1_na.bkg', overwrite=True)
```

```python
bkgrate = fits.open('CTA1_na.bkg')
bkgrate = Background3D.from_hdulist(bkgrate)
```

```python
for idx, obs in tqdm(enumerate(srcobs_na)):
    srcobs_na[idx] = obs.copy(in_memory=True, events=select_gammas(obs), bkg=bkgrate)
```

```python
srcobs_na[0].peek()
```

```python
mimic_na.find_mimic_runs()
```

```python
plot_runs_info(mimic_na.mimic_datasets[1]['information'],elmin=40,elmax=60)
```

```python
mimic_na.run_blank_sky(1)
mimic_na.run_shifted_sky(1)
```

```python
for idx, obs in tqdm(enumerate(mimic_na.mimic_datasets[1]['observations'])):
    mimic_na.mimic_datasets[1]['observations'][idx] = obs.copy(
        in_memory=True, events=select_gammas(obs), bkg=bkgrate)
```

```python
skydir = srcobs_na[0].target_radec
energy_axis_true = srcobs_na[0].edisp.axes['energy_true']
geom = WcsGeom.create(
    binsz=config['skymap']['binsz'], frame='icrs', axes=[energy_axis], skydir=skydir, width=config['skymap']['width'])
geom_image = geom.to_image().to_cube([energy_axis.squash()])
exclusion_mask = ~geom_image.region_mask(CircleSkyRegion(skydir,config['exclusion_radius']*u.deg))
exclusion_mask.plot()
```

```python
map_dataset_maker = MapDatasetMaker()
safe_mask_maker = SafeMaskMaker(methods=['aeff-max','offset-max'],
                                aeff_percent=config['aeff_max'], offset_max=config['offset_max']*u.deg)
datasets_maker = DatasetsMaker([map_dataset_maker, safe_mask_maker],
                               stack_datasets=False, n_jobs=config['parallel_njobs'],
                               cutout_mode='trim', cutout_width=config['offset_max']*2*u.deg)
```

```python
fov_bkg_maker = FoVBackgroundMaker(method='scale', exclusion_mask=exclusion_mask)
```

```python
stacked = MapDataset.create(geom, energy_axis_true=energy_axis_true, name='stacked')
datasets = datasets_maker.run(stacked, mimic_na.mimic_datasets[1]['observations'])
```

```python
for data in tqdm(datasets):
    dataset = fov_bkg_maker.run(data)
    stacked.stack(dataset)
```

```python
plot_fov([dataset.models[0].spectral_model.norm.value for dataset in datasets], 'mimic0')
```

```python
sig = sigmap(stacked,config['skymap']['binsz']*u.deg)
plot_sig(sig['sqrt_ts'], exclusion_mask, config['exclusion_radius']*u.deg, significance_map_conv=sig['sqrt_ts_conv'],
        vmin=-3, vmax=3)
```

```python
stacked.write('mimic0_na_dataset.fits')
```
